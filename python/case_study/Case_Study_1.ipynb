**1. DATA CLEANING**

**Reseting Columns**

```python
col = ["Roll_No","Student_Name","Gender","Age","Course","Subject","Email","Phone","Admission_Date","Attendance_%","GPA","Status" ]
df = pd.read_csv('students_record.csv', header = None, names = col)     #header = none -> there will be no old header, the old headers will be shifted one cell down
df.head()
```

```python
df.drop(index=[0,1], inplace=True)
df.head()
```

**Resetting Index**

```python
df.set_index("Roll_No", inplace = True)
df.head()
```

**Handling Missing Values**

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

```python
#as the age dtype is object we have to change it to int first so later we can replace the null values
df.replace("NaN", np.nan, inplace=True)
df.replace("Na", np.nan, inplace=True)
df.replace("inf", np.nan, inplace=True)
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')

# Fill NaN values in 'Age' with the mean before converting to integer
df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Age"] = df["Age"].astype(int)
df.head()
```

```python
#as the attendance dtype is object we have to change it to int first so later we can replace the null values
df.replace("NaN", np.nan, inplace=True)
df.replace("Na", np.nan, inplace=True)
df.replace("inf", np.nan, inplace=True)
df["Attendance_%"] = pd.to_numeric(df["Attendance_%"], errors='coerce')

# Fill NaN values in 'Age' with the mean before converting to integer
df["Attendance_%"] = df["Attendance_%"].fillna(df["Attendance_%"].mean())
df["Attendance_%"] = df["Attendance_%"].astype(float)
df["Attendance_%"] = df["Attendance_%"].round(2)
df.head()
```

```python
df["Subject"] = df["Subject"].fillna(df["Subject"].mode()[0])
df.head()
```
