```python
area = [600,800,1000,1200,1500]
price = [15,18,25,35,40]
```

```python
plt.plot(area,price)

plt.title("Area vs Price")
plt.xlabel("Area in sq.ft")
plt.ylabel("Price in Lakhs")
plt.grid(True)
plt.show()
```

xlim and ylim to define x and y range/limit

```python
plt.plot(area,price)

plt.title("Area vs Price")
plt.xlabel("Area in sq.ft")
plt.ylabel("Price in Lakhs")


plt.xlim(400,2000)
plt.ylim(10,50)

plt.grid(True)
plt.show()
```

xticks and yticks to rotate labels

```python
plt.plot(area,price)

plt.title("Area vs Price")
plt.xlabel("Area in sq.ft")
plt.ylabel("Price in Lakhs")


plt.xticks(rotation=45)
plt.yticks(rotation=-270)

plt.grid(True)
plt.show()
```
