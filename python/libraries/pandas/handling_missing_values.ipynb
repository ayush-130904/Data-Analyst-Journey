```python
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mdferojahmed178/12000-house-price-list-from-10-states-of-usa")

print("Path to dataset files:", path)
```

```python
import os
print("Files in the downloaded dataset directory:")
print(os.listdir(path))
df = pd.read_csv(path + "/US houuse price of 10 states.csv")
```

```python
df.isnull().sum()    #finding null values
```

```python
df.isnull().sum().sum()   #total null values in dataset
```

```python
df["house_size"] = df["house_size"].fillna(df["house_size"].mode()[0])       #handling missing values in the given column with mode method
```

```python
df.isnull().sum()
```

```python
df["bed"] = df["bed"].fillna(df["bed"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["bath"] = df["bath"].fillna(df["bath"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["broker"] = df["broker"].fillna("Unkown")     #fill nun values of broker with a string
```

```python
df.isnull().sum()
```

```python
df.to_csv("Cleaned_data.csv")
```
