**3. Setting New Indexes**

```python
col = ["Sr No","City","Coaching Centre","Courses","State","Course Duration"]
df = pd.read_csv('course.csv', header = None, names = col)     #header = none -> there will be no old header, the old headers will be shifted one cell down
df.head()
```

```python
df.set_index("Courses", inplace = True)
df.head()
```

Reseting the Index

```python
df.index.name = "Courses"
df.reset_index(inplace=True)
df.head()
```
