**Filtering**

```python
df.head()
```

```python
bx = []                                # filtering movies whose duration is more than or equal to 200

for length in df.duration:
    if length >= 200:
        bx.append(True)
    else:
        bx.append(False)
```

```python
bx[:10]
```

```python
sbx = pd.Series(bx)
df[sbx].head()
```

```python
mask = df["duration"]>=100
df[mask].head()                                 #masking/filtering for movies based on duration greater than or equal to 100
```

```python
df[df.duration>=200]["title"].head()               #filtering based on duration >=200 only titles
```

```python
df[df.duration>=200].genre.head()               #filtering based on duration >=200 only genre names
```

```python
df[(df.duration>=200) & (df.genre == "Crime")]               #filtering based on CONDITIONS -> duration >=200 & genre == Crime
```

```python
df[(df.duration>=200) | (df.genre == "Crime")].head()        #filtering based on CONDITIONS -> duration >=200 or genre == Crime
```

```python
df[(df.genre == "Crime") | (df.genre == "Drama") | (df.genre == "Action")].head()
```

```python
df[df.genre.isin(["Crime","Drama","Action"])].head()                  #.isin() method checks if each element in the df.genre Series is present in the provided list of genres.
```
