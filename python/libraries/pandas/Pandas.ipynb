```python
import pandas as pd
import numpy as np

import warnings                        #to ignore the warnings while running the program
warnings.filterwarnings('ignore')
```

**1. Series**

Series in 1D labelled array that is capable of holding data of any type

syntax -

pd.Series(data=[], index=[])

```python
s = pd.Series()
s
```

```python
s = pd.Series(data = [1,2,3], index = ['a','b','c'])
print(s)
```

```python
d = {'a':1, 'b':2, 'c':3}
s = pd.Series(data = d)
print(s)
```

```python
d = {'a':1, 'b':2, 'c':3}

s = pd.Series(data = d, index = ['c','a','d','b'])
print(s)
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array(["a","b","c"])

s = pd.Series(data = arr1, index = arr2)
print(s)
```

```python
s = pd.Series(data = 5, index = ['a','b','c'])    #if data is assigned single value then it is assigned to every index
print(s)
```

**2. DataFrame**

DataFrame is 2D data structure i.e data is aligned in tabular format which is rows and col

```python
df = pd.DataFrame()
df
```

```python
data = [1,2,3,4,5]
df = pd.DataFrame(data)
df
```

```python
data = [['1','2','3'],['alpha','beta','gamma'],['a','b','c']]

df = pd.DataFrame(data, columns = ["C1","C2","C3"])
df
```

```python
name = ['Pratik','Purvesh','Faizan','Sandesh','Madhur', 'Vinayak']
py = [89,76,54,78,98,87]
ml = [54,76,89,54,90,76]

dic = {"Name":name, "Python":py, "Machine Learning":ml}
df = pd.DataFrame(dic)
df
```

```python
df[df.Name=='Pratik']      #masking
```

```python
name = ['Pratik','Purvesh','Faizan','Sandesh','Madhur', 'Vinayak']
py = [89,76,54,78,98,87]
ml = [54,76,89,54,90,76]

dic = {"Name":name, "Python":py, "Machine Learning":ml}
df = pd.DataFrame(dic, index = ["rank1","rank2","rank3","rank4","rank5","rank6",])       #assigning index
df
```

```python
df = pd.DataFrame(np.arange(1,5).reshape(2,2), columns= ["A","B"])       #creating dataframe using np.arange
df
```

```python
d = {
    "one": pd.Series([1,2,3], index = ['a','b','c']),
    "two": pd.Series([1,2,3,4], index = ['a','b','c','d'])

    }

df= pd.DataFrame(d)
df
```
