**Masking/Filtering/Selection**

```python
#example1
x>5   #condition
```

```python
mask = x>5         #masking
x[mask]
```

```python
#or
x[x>5]
```

```python
#example2
names = np.array(['Pratik', 'Faizan', 'Alok', 'Pratik', 'Ayush', 'Pratik', 'Faizan'])
```

```python
names == 'Pratik'    #condition
mask = names == 'Pratik' #masking
names[mask]
```

```python
#or
names[names == 'Pratik']
```

```python
#select all values which is not 'Pratik'
names[names != 'Pratik']
```

OR -> |

```python
(names == 'Pratik') | (names == 'Faizan')   #condition
```

```python
names[(names == 'Pratik') | (names == 'Faizan')]  #masking
```

```python
#select all the values, which is not pratik and change them to purvesh
names[names != 'Pratik'] == 'Purvesh'
```

```python
#The dot product is the standard matrix multiplication you likely learned in linear algebra. It involves a "row-by-column" summation.
#Logic: For 2D matrices, the element at position $(i, j)$ in the output is the sum of the products of the $i$-th row of $a$ and the $j$-th column of $b$.
np.dot(a,b)     #dot product of a and b
```

```python
#While np.dot is optimized for 1D and 2D arrays (vectors and matrices), tensordot is the "heavy-duty" version designed to handle multidimensional arrays (tensors) by summing products along specific axes.
np.tensordot(a,b)
```

```python
#The Kronecker product (often denoted by the symbol ) is an operation that results in a block matrix. It is much more expansive than a dot product.
#Logic: Every single element in array $a$ is multiplied by the entire array $b$.
np.kron(a,b)
```

**Chapter 5 - Numpy - Statistical Operation**

```python
a = np.arange(1,6)
b = np.sqrt(a)
b
```

```python
#In NumPy, np.ceil(b) stands for the ceiling of the array elements. It is a mathematical operation that rounds each number up to the nearest integer.
#Positive numbers: It moves toward the next higher integer (e.g., 1.2 becomes 2.0).
#Negative numbers: It moves toward zero (e.g., -1.8 becomes -1.0), because $-1$ is mathematically "greater" than $-1.8$.
#Integers: If the number is already an integer (e.g., 3.0), it remains unchanged
np.ceil(b)
```

Eculedian Distance Formula between two vectors
$$d = \sqrt{\sum_{i=1}^{n} (v1_i - v2_i)^2}$$

```python
v1 = np.random.randn(10)
v2 = np.random.randn(10)
```

```python
v1
```

```python
v2
```

```python
dist = np.sqrt(np.sum((v1-v2)**2))
```
