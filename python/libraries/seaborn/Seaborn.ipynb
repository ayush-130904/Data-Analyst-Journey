```python
sns.distplot(df['price'])
```

```python
mwidth = df['width'].mean()
dwidth = df['width'].median()

sns.distplot(df['width'])
plt.axvline(mwidth, color = "red")
plt.axvline(dwidth, color = "yellow")
plt.grid(True)
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data = df, x = "body-style", hue = "fuel-type")
plt.grid(True)
```

```python
df.replace("?", np.nan, inplace=True)

df['normalized-losses'] = pd.to_numeric(df['normalized-losses'], errors='coerce')
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Blues")
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="viridis")
```

**Subplots**

```python
x = [1,2,3,4]
y = [3,4,5,6]

fig, axes = plt.subplots(2,2, figsize=(8,6))
axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].plot(x,y)
axes[1,1].scatter(x,y)
```

```python
fig, axes = plt.subplots(2,2, figsize=(20,10))

sns.distplot(df["price"], ax=axes[0,0])
sns.countplot(data=df, y = 'make', ax=axes[0,1])
sns.scatterplot(data=df, x='width', y = 'price', hue= "body-style", ax=axes[1,0])
sns.boxplot(data=df, x = 'price', ax=axes[1,1])
```

```python
df = pd.read_csv("/content/Data-Analyst-Journey/python/libraries/seaborn/gas_prices.csv")
df.head()
```

```python
plt.plot(df.Year, df.USA, 'b+-', label="United States" )
plt.plot(df.Year, df['South Korea'], 'r.-', label="South Korea" )
plt.plot(df.Year, df.Canada, 'g*-', label="Canada" )

plt.title("Gas Used By Countries per Year", fontdict={"fontname":"Liberation Mono",
 "fontsize":20})

plt.xlabel("Per Year")
plt.ylabel("Gas used")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.font_manager as fm

# Get a list of all available font families
font_families = [f.name for f in fm.fontManager.ttflist]

# Print the unique font families
print("Available font families in Matplotlib:")
for font_name in sorted(list(set(font_families))):
    print(f"- {font_name}")
```
